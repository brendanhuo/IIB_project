```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

path = 'Blast_conservation/7NKFSUNH016-Alignment-HitTable.csv'
```

```python
blast_select = pd.read_csv(path, header = None)
selectivity = []
for i in range(127):
    selectivity.append(1-np.mean(blast_select[blast_select.iloc[:,0] == 'gRNA{}'.format(i)].iloc[:,1])/20)
```

```python

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(np.arange(len(selectivity)), selectivity, width = 0.6, color = 'r')
plt.xlabel('gRNA sequence');plt.ylabel('Selectivity scores');plt.grid()
plt.ylim(0.3, 0.4), plt.xlim(-1, len(selectivity))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(np.arange(len(selectivity)), np.sort(selectivity), width = 0.6, color = 'r')
plt.xlabel('gRNA sequence');plt.ylabel('Selectivity scores');plt.grid()
plt.ylim(0.3, 0.4), plt.xlim(-1, len(selectivity))
plt.show()
```
